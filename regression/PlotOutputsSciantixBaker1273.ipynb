```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 14})
```

```python
# Leggi i dati da file
data1 = pd.read_csv('test_Baker1977__1273K/output.txt', delimiter='\t')

data2 = pd.read_csv('test_Baker1977__1273K/outputPVU.txt', delimiter='\t')

data3 = pd.read_csv('test_Baker1977__1273K/outputArianna.txt', delimiter='\t')
l = 3
data = [data1, data2, data3]
labels = ['No athermal', 'PVU', 'Arianna']
kinds = ['-', ':', '-.']
N = len(data[0])
print(N)

time_ref = 0

# Creazione di una nuova colonna "Time (min)" per il tempo relativo in minuti
for i in range(l):  # Itera su data1 e data2
    data[i]["Fission gas release (%)"] = (data[i]["Fission gas release (/)"]) * 100 

plt.rcParams.update({'font.size': 18})
plt.rcParams.update({'lines.markersize': 6})  # Corrected parameter for marker size
plt.rcParams.update({'lines.linewidth': 2}) 

fig, ax = plt.subplots(figsize=(11,8))

for i in range(l):
    data[i].plot(x="Burnup (MWd/kgUO2)", y="Fission gas release (%)", kind="line", linestyle=kinds[i], ax=ax, label=labels[i])

# Impostazioni dell'asse principale (Temperatura)
ax.set_xlabel("Burnup (MWd/kgUO2)") 
ax.set_ylabel("Fission Gas Release (%)")

# Impostazioni aggiuntive dell'asse x
plt.tight_layout()

fig, ax = plt.subplots(figsize=(11,8))

for i in range(l):
    data[i].plot(x="Burnup (MWd/kgUO2)", y="Intergranular fractional coverage (/)", kind="line", linestyle=kinds[i], ax=ax,label=labels[i])

# Impostazioni dell'asse principale (Temperatura)
ax.set_xlabel("Burnup (MWd/kgUO2)") 
ax.set_ylabel("Intergranular fractional coverage (/)")

# Impostazioni aggiuntive dell'asse x
plt.tight_layout()

# Creazione di una nuova colonna "Time (min)" per il tempo relativo in minuti
for i in [2]:  # Itera su data1 e data2
    data[i]["Porosity (%)"] = (data[i]["Porosity (/)"]) * 100 
    data[i]["Open porosity (%)"] = (data[i]["Open porosity (/)"]) * 1000
    data[i]["Fabrication porosity (%)"] = (data[i]["Fabrication porosity (/)"]) * 100

fig, ax = plt.subplots(figsize=(11,8))

for i in [2]:
    data[i].plot(x="Burnup (MWd/kgUO2)", y="Porosity (%)", kind="line", linestyle=kinds[i], ax=ax)
    data[i].plot(x="Burnup (MWd/kgUO2)", y="Fabrication porosity (%)", kind="line", linestyle=kinds[i], ax=ax)
    data[i].plot(x="Burnup (MWd/kgUO2)", y="Open porosity (%)", kind="line", linestyle=kinds[i], ax=ax, label='Open porosity(%) *10')

# Impostazioni dell'asse principale (Temperatura)
ax.set_xlabel("Burnup (MWd/kgUO2)") 
ax.set_ylabel("Porosity (%)")

# Impostazioni aggiuntive dell'asse x
plt.tight_layout()
```

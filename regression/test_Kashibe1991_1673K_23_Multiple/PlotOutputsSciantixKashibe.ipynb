## Data import

```python
dataEC0 = pd.read_csv('output.txt', delimiter='\t')
labelEC0= 'new'
dataold = pd.read_csv('output_gold.txt', delimiter='\t')
labelold= 'Barani 2017'
data = [dataEC0, dataold]
labels = [labelEC0, labelold]
N = len(data[0])
print(N)
```

## Bubble pressure

```python
xaxistoplot = 'Time (h)'
whattoplot= 'Intergranular bubble pressure (MPa)'
fig, ax = plt.subplots(1,1,figsize=(12, 4))
for i in range(len(labels)): 
    if whattoplot in data[i]:
        #data[i].plot(x=xaxistoplot, y='Equilibrium bubble pressure (MPa)', kind='line', label='Equilibrium', ax=ax)
        #data[i].plot(x=xaxistoplot, y='Hydrostatic stress (MPa)', kind='line', label='Hydrostatic stress', ax=ax)
        data[i].plot(x=xaxistoplot, y=whattoplot, kind='line', label=labels[i], ax=ax)
        data[i].plot(x=xaxistoplot, y='Critical intergranular bubble pressure (MPa)', kind='line', label='Critical', ax=ax)
ax.set_xlabel(xaxistoplot)
ax.set_ylabel(whattoplot)
ax.set_xlim(timeboundaries)
ax.set_title('Comparison')
ax.legend()
ax.grid(True)

fig, ax = plt.subplots(1,1,figsize=(12, 4))
for i in range(len(labels)): 
    if whattoplot in data[i]:
        data[i]['Overpressure']=(data[i]['Intergranular bubble pressure (MPa)']/data[i]['Critical intergranular bubble pressure (MPa)'])
        data[i].plot(x=xaxistoplot, y='Overpressure', kind='line', label='Ratio'+labels[i], ax=ax)

ax.set_xlabel(xaxistoplot)
ax.set_ylabel('P/Pc')
#ax.set_xlim(timeboundaries)
ax.set_title('Comparison')
ax.legend()
ax.grid(True)

fig, ax = plt.subplots(1,1,figsize=(12, 4))
for i in range(len(labels)): 
    if whattoplot in data[i]:
        data[i].plot(x=xaxistoplot, y='Intergranular fractional intactness (/)', kind='line', label='Ratio '+labels[i], ax=ax)

ax.set_xlabel(xaxistoplot)
ax.set_ylabel('Intergranular fractional intactness (/)')
#ax.set_ylim(0,6000)
ax.set_title('Comparison')
ax.legend()
ax.grid(True)
```

```python
xaxistoplot = 'Temperature (K)'
whattoplot= 'Intergranular bubble pressure (MPa)'
fig, ax = plt.subplots(1,1,figsize=(12, 4))
for i in range(len(labels)): 
    if whattoplot in data[i]:
        data[i].plot(x=xaxistoplot, y='Critical intergranular bubble pressure (MPa)', kind='line', label='Critical pressure'+labels[i], ax=ax)
        data[i].plot(x=xaxistoplot, y='Intergranular bubble pressure (MPa)', kind='line', label='Bubble pressure '+labels[i], ax=ax)
ax.set_xlabel(xaxistoplot)
ax.set_ylabel(whattoplot)
ax.set_title('Comparison')
ax.set_yscale('log')
ax.legend()
ax.grid(True)
```

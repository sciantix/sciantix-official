```python
import pandas as pd
import matplotlib.pyplot as plt

# Caricamento dei dati
database = pd.read_csv('test_Baker1977__1273K/output.txt', sep='\t')

# Creazione del plot
plt.figure(figsize=(8, 6))
plt.plot(database['Time (h)'], database['Cs at grain boundary (at/m3)'], label='Cs - gb', linestyle='-')
plt.plot(database['Time (h)'], database['Cs reacted (at/m3)'], label='Cs - reacted', linestyle='-')
plt.plot(database['Time (h)'], database['I at grain boundary (at/m3)'], label='I - gb', linestyle='-')
plt.plot(database['Time (h)'], database['I reacted (at/m3)'], label='I - reacted', linestyle='-')
plt.plot(database['Time (h)'], database['CsI produced (at/m3)'], label='CsI - produced', linestyle='-')
plt.plot(database['Time (h)'], database['CsI at grain boundary (at/m3)'], label='CsI - gb', linestyle='-')

# Label e titolo
plt.xlabel('Time (h)')
plt.ylabel('Concentration (at/m3)')
plt.title('Cs and CsI Evolution Over Time')
plt.legend()
plt.grid(True)

database['R/B Cs'] = database['Cs released (at/m3)']/database['Cs produced (at/m3)']
database['R/B I'] = database['I released (at/m3)']/database['I produced (at/m3)']
database['R/B CsI'] = database['CsI released (at/m3)']/database['CsI produced (at/m3)']


# Creazione del plot
plt.figure(figsize=(8, 6))
plt.plot(database['Time (h)'], database['R/B Cs'], label='Cs', linestyle='-')
plt.plot(database['Time (h)'], database['R/B I'], label='I', linestyle='-')
plt.plot(database['Time (h)'], database['R/B CsI'], label='CsI', linestyle='-')

# Label e titolo
plt.xlabel('Time (h)')
plt.ylabel('Concentration (at/m3)')
plt.title( 'Release')
plt.legend()
plt.grid(True)

```

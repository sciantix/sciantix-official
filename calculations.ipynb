```python
import numpy as np
import matplotlib.pyplot as plt
from pymatgen.analysis.phase_diagram import PhaseDiagram, PDPlotter, PDEntry

# Definizione delle funzioni di Gibbs (esempio)
def G_FeO(T):
    return -5.5 + 0.0005 * T  # eV/atomo

def G_Fe2O3(T):
    return -8.0 + 0.0007 * T  # eV/atomo

# Selezione della temperatura
T_target = 1000  # Kelvin

# Calcolo di G(T) per i composti a T_target
G_FeO_T = G_FeO(T_target)
G_Fe2O3_T = G_Fe2O3(T_target)

# Definizione delle fasi nel diagramma di fase
entries = [
    PDEntry("Fe", -8.3),      # Energia libera del ferro puro (ipotizzata)
    PDEntry("O2", 0),         # Energia dell'ossigeno molecolare (riferimento)
    PDEntry("FeO", G_FeO_T),  # Energia libera di FeO a T_target
    PDEntry("Fe2O3", G_Fe2O3_T)  # Energia libera di Feâ‚‚Oâ‚ƒ a T_target
]

# Costruzione del diagramma di fase
pd = PhaseDiagram(entries)
plotter = PDPlotter(pd)
title_font = dict(family="Arial", size=16, color="darkblue")
# Personalizzazione del titolo (in Plotly)
fig = plotter.get_plot()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definizione delle funzioni di Gibbs
def G_o(T):
    # Condizioni per T < 903 e T >= 903 usando np.where per gestire gli array
    A = np.where(T < 903, -533585, -523534)
    B = np.where(T < 903, 266.381, 419.454)
    C = np.where(T < 903, -3.953, -29.576)
    D = np.where(T < 903, -0.011305, 0.0000675)
    E = np.where(T < 903, 0, 0)
    return A + B*T + C*T*np.log(T) + D*T**2 + E/T

def K(T):
    return np.exp(-G_o(T)/8.314/T)

# Tracciare la funzione di Gibbs in funzione della temperatura
Trange = np.linspace(1000, 2000, 100)
x_I = np.linspace(0,1,100)
x_Cs = np.linspace(0,1,100)
x_CsI = 1 - x_I - x_Cs
plt.plot(Trange, K(Trange))
plt.xlabel("Temperatura (K)")
plt.ylabel("Energia libera di Gibbs (G_o)")
plt.title("Funzione di Gibbs in funzione della temperatura")
plt.grid(True)
plt.show()
```
